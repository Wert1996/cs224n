```python
with open('adultsaretalking.txt', 'r') as f:
    text = f.read()
```

```python
import re

def get_distinct_words(text: str):
    words = re.findall(r'\b[a-zA-Z]+\b', text.lower())
    return list(set(words))
```

```python
import numpy as np

def get_cooccurence_matrix(text: str, window_size: int):
    distinct_words = get_distinct_words(text)
    word_to_index = {word: i for i, word in enumerate(distinct_words)}
    cooccurence_matrix = np.zeros((len(distinct_words), len(distinct_words)))
    words = re.findall(r'\b[a-zA-Z]+\b', text.lower())
    for i in range(len(words)):
        if i == len(words) - window_size:
            break
        center_word = words[i]
        center_word_ind = word_to_index[center_word]
        for word in words[i+1:i+window_size]:
            neighbor_word_ind = word_to_index[word]
            cooccurence_matrix[center_word_ind][neighbor_word_ind] += 1
            cooccurence_matrix[neighbor_word_ind][center_word_ind] += 1
    return cooccurence_matrix
```

```python
cmat = get_cooccurence_matrix(text, 4)
dwords = get_distinct_words(text)
```

```python
from sklearn.decomposition import TruncatedSVD

def reduce_to_k_dim(M: np.array, k=2):
    svd = TruncatedSVD(n_components=k, n_iter=10, random_state=42)
    return svd.fit_transform(M)
```

```python
reduced_cmat = reduce_to_k_dim(cmat, k=100)
```
